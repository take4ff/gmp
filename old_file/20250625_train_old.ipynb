```python
def extract_keys_in_range(data, start_key, end_key):
    return [value for key, value in data.items() if start_key <= key <= end_key]

def separate_train_test(data, test_start=30, ylen=1):
    train_data_x = []
    train_data_y = []
    test_data_x = {}
    test_data_y = {}

    for d in data:
        if len(d) < test_start:
            train_data_x.append(extract_keys_in_range(d, 1, len(d) - ylen))
            train_data_y.append(extract_keys_in_range(d, len(d) - ylen + 1, len(d)))
        else:
            train_data_x.append(extract_keys_in_range(d, 1, test_start - 1))
            train_data_y.append(extract_keys_in_range(d, test_start, test_start + ylen - 1))

            for i in range(test_start, len(d) + 1 - ylen + 1):
                if i not in test_data_x:
                    test_data_x[i] = []
                    test_data_y[i] = []
                

                test_data_x[i].append(extract_keys_in_range(d, i - test_start + 1, i - 1))
                test_data_y[i].append(extract_keys_in_range(d, i, i + ylen - 1))

    return train_data_x, train_data_y, test_data_x, test_data_y
```

```python
train_x, train_y, test_x, test_y = separate_train_test(data, test_start=30,ylen=1)
```

```python
def pick_up_feature(paths, feature_name):
    features = []
    for path in paths:
        feature = []
        for mutations in path:
            for mutation in mutations:
                if feature_name in mutation:
                    feature.append(mutation[feature_name])
        features.append(feature)
    return features

train_y_protein = pick_up_feature(train_y, 'protein')
print("train_y_protein:", len(train_y_protein), train_y_protein)
test_y_protein = {}
for i in range(30, 30+len(test_y)):
    if i not in test_y_protein:
        test_y_protein[i] = []
    test_y_protein[i] = (pick_up_feature(test_y[i], 'protein'))
print("test_y_protein:", len(test_y_protein), test_y_protein)
```

```python
import torch
import torch.nn as nn

# 下位モデル: タイムステップ内エンコーダ
class LocalEncoder(nn.Module):
    def __init__(self, input_dim, embed_dim, num_heads, ff_dim, dropout=0.1):
        super(LocalEncoder, self).__init__()
        self.embedding = nn.Linear(input_dim, embed_dim)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(
                d_model=embed_dim,
                nhead=num_heads,
                dim_feedforward=ff_dim,
                dropout=dropout
            ),
            num_layers=1
        )
        self.pooling = nn.AdaptiveAvgPool1d(1)  # タイムステップ内の要約

    def forward(self, x):
        # x: (batch_size, num_features, input_dim)
        x = self.embedding(x)  # (batch_size, num_features, embed_dim)
        x = x.permute(1, 0, 2)  # (num_features, batch_size, embed_dim)
        x = self.transformer(x)  # (num_features, batch_size, embed_dim)
        x = x.permute(1, 2, 0)  # (batch_size, embed_dim, num_features)
        x = self.pooling(x).squeeze(-1)  # (batch_size, embed_dim)
        return x

# 上位モデル: タイムステップ間エンコーダ
class GlobalEncoder(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, num_layers, dropout=0.1):
        super(GlobalEncoder, self).__init__()
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(
                d_model=embed_dim,
                nhead=num_heads,
                dim_feedforward=ff_dim,
                dropout=dropout
            ),
            num_layers=num_layers
        )

    def forward(self, x):
        # x: (batch_size, num_timesteps, embed_dim)
        x = x.permute(1, 0, 2)  # (num_timesteps, batch_size, embed_dim)
        x = self.transformer(x)  # (num_timesteps, batch_size, embed_dim)
        x = x.permute(1, 0, 2)  # (batch_size, num_timesteps, embed_dim)
        return x

# 全体モデル: 下位モデルと上位モデルの統合
class HierarchicalTransformer(nn.Module):
    def __init__(self, input_dim, embed_dim, num_heads, ff_dim, num_layers, dropout=0.1):
        super(HierarchicalTransformer, self).__init__()
        self.local_encoder = LocalEncoder(input_dim, embed_dim, num_heads, ff_dim, dropout)
        self.global_encoder = GlobalEncoder(embed_dim, num_heads, ff_dim, num_layers, dropout)
        self.classifier = nn.Linear(embed_dim, 1)  # 最終的な予測タスク用

    def forward(self, x):
        # x: (batch_size, num_timesteps, num_features, input_dim)
        batch_size, num_timesteps, num_features, input_dim = x.size()
        x = x.view(batch_size * num_timesteps, num_features, input_dim)  # (batch_size * num_timesteps, num_features, input_dim)
        local_features = self.local_encoder(x)  # (batch_size * num_timesteps, embed_dim)
        local_features = local_features.view(batch_size, num_timesteps, -1)  # (batch_size, num_timesteps, embed_dim)
        global_features = self.global_encoder(local_features)  # (batch_size, num_timesteps, embed_dim)
        output = self.classifier(global_features[:, -1, :])  # 最後のタイムステップの特徴量を使用
        return output

# モデルの初期化
input_dim = 10  # 入力次元（例: 各変異の特徴量数）
embed_dim = 64  # 埋め込み次元
num_heads = 4  # マルチヘッドアテンションのヘッド数
ff_dim = 128  # フィードフォワード層の次元
num_layers = 2  # 上位モデルの層数
dropout = 0.1  # ドロップアウト率

model = HierarchicalTransformer(input_dim, embed_dim, num_heads, ff_dim, num_layers, dropout)
print(model)
```

```python
def convert_dict_to_list(data):
    """
    Convert a nested dictionary structure to a list format.
    Args:
        data (list of dict): Input data where each element is a dictionary.
    Returns:
        list: Converted data in list format.
    """
    converted_data = []
    for d in data:
        if isinstance(d, dict):
            # Flatten the dictionary values into a list
            flattened = []
            for key in sorted(d.keys()):
                flattened.extend(d[key])
            converted_data.append(flattened)
        else:
            raise TypeError(f"Unsupported data type: {type(d)}")
    return converted_data

# Example usage
train_x_list = convert_dict_to_list(train_x)
print("Converted train_x to list format:", train_x_list)
```

```python
from torch.utils.data import DataLoader, TensorDataset

# データをテンソルに変換
train_x_tensor = torch.tensor(train_x, dtype=torch.float32)
train_y_tensor = torch.tensor(train_y_protein, dtype=torch.float32)

# データローダーの作成
batch_size = 32
train_dataset = TensorDataset(train_x_tensor, train_y_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# 損失関数と最適化手法
criterion = nn.MSELoss()  # 例: 回帰タスクの場合
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# トレーニングループ
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch_x, batch_y in train_loader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)
        
        # 順伝播
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        
        # 逆伝播と最適化
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}")

# モデルの評価
model.eval()
test_x_tensor = torch.tensor(test_x, dtype=torch.float32).to(device)
test_y_tensor = torch.tensor(test_y_protein, dtype=torch.float32).to(device)

with torch.no_grad():
    predictions = model(test_x_tensor)
    test_loss = criterion(predictions, test_y_tensor)
    print(f"Test Loss: {test_loss.item():.4f}")

import matplotlib.pyplot as plt

# 予測と実際の値をプロット
plt.plot(predictions.cpu().numpy(), label="Predictions")
plt.plot(test_y_tensor.cpu().numpy(), label="Actual")
plt.legend()
plt.show()

# モデルの保存
torch.save(model.state_dict(), "hierarchical_transformer.pth")
print("Model saved successfully.")
```
